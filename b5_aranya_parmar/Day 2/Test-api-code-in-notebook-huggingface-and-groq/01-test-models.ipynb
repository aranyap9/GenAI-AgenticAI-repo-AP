```python
import os
from openai import OpenAI
```

###### Alternatively, we can use python's requests module for sending the HTTP request to the HF server

```python
client = OpenAI(
    base_url="https://router.huggingface.co/v1",
    api_key=api_key
)
client
```

```python
response = client.chat.completions.create(
    model="Qwen/Qwen2.5-72B-Instruct:fastest",
    messages=[
        {"role": "system", "content":"You are a helpful and brief AI."},
        {"role": "user", "content": "Tell me 5 facts about Voyager 1"}
    ],
    max_tokens=500,
)
```

```python
print(response.choices[0].message.content)
```
